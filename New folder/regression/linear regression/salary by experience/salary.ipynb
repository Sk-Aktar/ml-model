```python
import numpy as np
import pandas as pd
# Create a simple dataset with one independent variable
np.random.seed(42)
n = 100

# Independent variable: Years of Experience
experience = np.random.uniform(0, 10, n).round(2)

# Dependent variable: Salary (in ₹1000s)
salary = (experience * 6.5 + np.random.normal(0, 5, n)).round(2)  # Linear relation + noise

# Create DataFrame
df= pd.DataFrame({
    'Experience': experience,
    'Salary': salary
})

#tools.display_dataframe_to_user(name="Single Variable Linear Regression Dataset", dataframe=df_single)
df.head()
```

```python
df.describe()
```

```python
x= df[['Experience']]
y = df['Salary']
```

```python
import matplotlib.pyplot as plt
# Visualize the dataset
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Salary vs Experience')
#plt.xlabel('Years of Experience')
#plt.ylabel('Salary (in ₹1000s)')
#plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```
