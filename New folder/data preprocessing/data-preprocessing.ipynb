1. data import and inspect

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eva', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack'],
    'Age': [20, 21, None, 23, 22, 25, 20, 21, 19, 30],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Marks': [85, 78, 90, None, 45, 55, 88, 60, 95, 35],
    'Attendance (%)': [92, 88, 95, 70, 60, 50, None, 75, 98, 40],
    'Study Hours': [5.5, 4.0, 6.0, 2.0, 1.0, 1.5, 5.0, 3.0, 6.5, 0.5],
    'Passed': ['Yes', 'Yes', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Country': ['India', 'India', 'Canada', 'Canada', 'India', 'UK', 'UK', 'India', 'Canada', 'UK']
}
df = pd.DataFrame(data)
```

```python
df.head()  # Display the first few rows of the DataFrame
```

```python
df.info()  # Display information about the DataFrame
```

```python
df.describe()  # Get a statistical summary of the DataFrame
```

2. handling missing value 
    a. check for missing value

```python
df.isnull().sum()  # Check for missing values in the DataFrame
```

```python
df.isnull().any()  # Check if there are any missing values in the DataFrame
```

```python
df.isnull().sum().sum() # Count total missing values in the DataFrame
```

b. visualizing missing value

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull() , cmap='viridis', cbar=False)
plt.show()
```

```python
df
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)  # Fill missing Age with mean
df['Marks'].fillna(df['Marks'].mean(), inplace=True)  # Fill missing Marks with mean
df['Attendance (%)'].fillna(df['Attendance (%)'].mean(), inplace=True)  # Fill missing Attendance with mean
df
```

```python
df.isnull().sum()  # Check for missing values again after filling
```
