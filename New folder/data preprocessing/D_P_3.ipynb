```python
import pandas as pd
df=pd.read_csv('extended_big_dataset.csv')
df.head(30)  # Display the first 100 rows of the dataset
```

```python
df.isnull().sum()  # Check for missing values in the dataset
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)  # Fill missing Age with mean
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True) # Fill missing
df['City'].fillna(df['City'].mode()[0], inplace=True)  # Fill missing City with mode
df['Salary'].fillna(df['Salary'].interpolate(method='linear'), inplace=True)  # Fill missing Salary with mean
df['Education'].fillna(df['Education'].mode()[0], inplace=True)  # Fill missing Rating with mean
df['Experience'].fillna(df['Experience'].median(), inplace=True)  # Fill missing Last_Visit with mode
df['Department'].fillna(df['Department'].mode()[0], inplace=True)  # Fill missing Last_Visit with mode
df['Rating'].fillna(df['Rating'].median(), inplace=True)  # Fill missing Rating with mean
df['Join_Date'].fillna(df['Join_Date'].ffill(), inplace=True)  # Fill missing Last_Visit with mode
df['Last_Promotion_Year'].fillna(df['Last_Promotion_Year'].bfill(), inplace=True)  # Fill missing Last_Visit with mode
```

```python
df.isnull().sum()  # Check for missing values in the dataset
```
