```python
docs['label'] = docs.Class.map({'ham': 0, 'spam': 1})
```

```python
x = docs.sms
y = docs.label
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=1)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
data= ['the quick brown fox jumps over the lazy dog']
vectorizer = CountVectorizer()
vectorizer.fit(data)
print('given data:', data)
print("-"*20)
print('vocabulary:', vectorizer.vocabulary_)
print("-"*20)
vector = vectorizer.transform(data)
print(vector.shape)
print(type(vector))
print("-"*20)
print(vectorizer.get_feature_names_out())
print(vector.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
vect.fit(x_train)
```

```python
vect.vocabulary_
```

```python
vect.get_feature_names_out()
```

```python
x_train_tran = vect.transform(x_train)
x_test_tran = vect.transform(x_test)
```

```python
type(x_train_tran)
```

```python
print(x_train_tran.shape)
```

```python
print(x_train_tran[1, :].toarray())
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train_tran, y_train)
y_pred_class = bnb.predict(x_test_tran[1, :].toarray())
print(y_pred_class)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train_tran, y_train)
y_pred_class = bnb.predict(x_test_tran)
print(y_pred_class)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
bnb
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred_class)
print(confusion)
tn = confusion[0, 0]
fp = confusion[0, 1]
fn = confusion[1, 0]
tp = confusion[1, 1]
```

```python
sensitivity = tp / float(tp + fn)
print("Sensitivity:", sensitivity)
```

```python
specificity = tn / float(tn + fp)
print("Specificity:", specificity)
```

```python
precision = tp / float(tp + fp)
print("Precision:", precision)
print(metrics.precision_score(y_test, y_pred_class))
```

```python
print("Recall:", metrics.recall_score(y_test, y_pred_class))
print("F1 Score:", metrics.f1_score(y_test, y_pred_class))
print("precision:", metrics.precision_score(y_test, y_pred_class))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_class))
```
