```python
import pandas as pd
df=pd.read_csv("naive_bayes_dataset2.csv")
df.head()
```

```python
df=pd.get_dummies(df, columns=['Contains_Promotion_Words'], drop_first=True,dtype=int)
df=pd.get_dummies(df, columns=['Sender_Reputation'], drop_first=True,dtype=int)

df
```

```python
x=df.drop("Spam", axis=1)
y=df["Spam"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc= sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train_sc, y_train)
y_pred_class = model.predict(x_test_sc)
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
print("F1 Score:", f1_score(y_test, y_pred_class))
print("Accuracy Score:", accuracy_score(y_test, y_pred_class))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_class))
```
